```python
class node(object):
    def __init__(self,value):
        self.info = value
        self.next = None
        self.prev = None
```

```python
class DoubleLinkedList(object):
    def __init__(self):
        self.start = None
    
    def display_list(self):
        if self.start is None:           #list is empty
            print("List is empty")
            return
        p = self.start
        while p is not None:
            print(p.info," ",end=" ")
            p = p.next
        print()
        
    def insert_in_beginning(self,data):  #Inserting at the first position
        temp = node(data)
        temp.next = self.start
        self.start.prev = temp
        self.start = temp
        
    def insert_in_empty_list(self,data):
        temp = node(data)
        self.start = temp

    def insert_in_end(self,data):
        temp = node(data)
        p = self.start
        while p.next is not None:
            p = p.next
        p.next = temp
        temp.prev = p
        
    def create_list(self):
        n = int(input("Enter the number of nodes: "))
        if n==0:
            return
        data = int(input("Enter the first element to be inserted: "))
        self.insert_in_empty_list(data)
        
        for i in range(n-1):
            data = int(input("Enter the next element to be inserted: "))
            self.insert_in_end(data)
            
    def insert_after(self,data,x):
        temp = node(data)
        p = self.start
        while p is not None:     
            if p.info == x:      #Finding the position to insert the element
                break
            p = p.next
            
        if p is None:
            print(x,"is not present in the list")
        else:
            temp.prev = p
            temp.next = p.next    #Inserting the element
            if p.next is not None:
                p.next.prev = temp
            p.next = temp
            
    def insert_before(self,data,x):
        if self.start is None:
            print("List is empty")
            
        if self.start.info == x:  #Inserting element after the beginning
            temp = node(data)
            temp.next = self.start
            self.start.prev = temp
            self.start = temp
            return
        
        temp = node(data)
        p = self.start
        while p is not None:
            if p.info == x:
                break
            p = p.next
            
        if p is None:
            print(x,"is not present in the list")
        else:
            temp.next = p
            temp.prev = p.prev
            p.prev.next = temp
            p.prev = temp
    
    def delete_first_node(self):
        if self.start is None:         #list is empty
            return
        if self.start.next is None:    #list has only one node
            self.start = None
            return
        self.start = self.start.next
        self.start.prev = None
        
    def delete_last_node(self):
        if self.start is None:         #list is empty
            return
        if self.start.next is None:    #list has only one node
            self.start = None
            return
        p = self.start
        while p.next is not None:
            p = p.next
        p.prev.next = None
        
    def delete_node(self,x):
        if self.start is None:          #List is empty
            return
        if self.start.next is None:     #List has only one node
            if self.start.info == x:
                self.start = None
            else:
                print(x," not found")
            return
            
        if self.start.info == x:       #First element is to be deleted
            self.start = self.start.next
            self.start.prev = None
            return
        
        p = self.start.next            #Finding the element to be deleted
        while p.next is not None:
            if p.info == x:
                break
            p = p.next
            
        if p.next is not None:          #Deleting the element
            p.prev.next = p.next
            p.next.prev = p.prev
        else:
            if p.info == x:
                p.prev.next = None
            else:
                print(x," not found")
                
    def reverse_list(self):
        if self.start is None:
            return
        
        p1 = self.start
        p2 = p1.next
        p1.next = None
        p1.prev = p2
        while p2 is not None:
            p2.prev = p2.next
            p2.next = p1
            p1 = p2
            p2 = p2.prev
        self.start = p1      
```

```python
list = DoubleLinkedList()
print()
while True:
    print("1.Create list")
    print("2.Display list")
    print("3.Insert node in an empty list")
    print("4.Insert node at the beginning of a list")
    print("5.Insert node at the end of the list")
    print("6.Insert node after a specified node")
    print("7.Insert node before a specified node")
    print("8.Delete first node")
    print("9.Delete last node")
    print("10.Delete any node")
    print("11.Reverse the list")
    print("12.Quit")
    print()
    
    option = int(input("Enter a choice: "))
    
    if option == 1:
        list.create_list()
    elif option == 2:
        list.display_list()
    elif option == 3:
        data = int(input("Enter the element to be inserted: "))
        list.insert_in_empty_list(data)
    elif option == 4:
        data = int(input("Enter the element to be inserted: "))
        list.insert_in_beginning(data)
    elif option == 5:
        data = int(input("Enter the element to be inserted: "))
        list.insert_in_end(data)
    elif option == 6:
        data = int(input("Enter the element to be inserted: "))
        x = int(input("Enter the element after which to insert: "))
        list.insert_after(data,x)
    elif option == 7:
        data = int(input("Enter the element to be inserted: "))
        x = int(input("Enter the element before which to insert: "))
        list.insert_before(data,x)
    elif option == 8:
        list.delete_first_node()
    elif option == 9:
        list.delete_last_node()
    elif option == 10:
        x = int(input("Enter the element to be deleted: "))
        list.delete_node(x)
    elif option == 11:
        list.reverse_list()
    elif option == 12:
        break
    else:
        print("Wrong option")
    print()
```
