```python
class node:
    def __init__(self,value):
        self.info = value
        self.link = None
```

```python
class SingleLinkedList:
    
    def __init__(self):
        self.start = None
    
    def display_list(self):
        if self.start is None:
            print("List is empty")
            return
        else:
            p = self.start
            while p is not None:
                print(p.info," ",end=" ")
                p = p.link
            print()
            
    def count_nodes(self):
        p = self.start
        n = 0
        while p is not None:
            n = n + 1
            p = p.link
        print("Number of nodes is: ",n)
        
    def search(self,x):
        pos = 1
        p = self.start
        while p is not None:
            if p.info == x:
                print(x,"is in position ",pos)
                return True
            pos += 1
            p = p.link
        else:
            print(x,"is not found")
            return False

##################INSERTION##########################       
   
    def insert_at_beginning(self,data):
        temp = node(data)
        temp.link = self.start
        self.start = temp
    
    def insert_at_end(self,data):
        temp = node(data)
        if self.start is None:
            self.start = temp
            return
        p = self.start
        while p.link is not None:
            p = p.link
        p.link = temp
        
    def create_list(self):
        n = int(input("Enter the no. of nodes:"))
        if n==0:
            return
        for i in range(n):
            data = int(input('Enter the data to be inserted:'))
            self.insert_at_end(data)
            
    def insert_after(self,data,x):
        p = self.start
        while p is not None:
            if p.info == x:
                break
            p = p.link
            
        if x is None:
            print(x,'is not present')
        else:
            temp = node(data)
            temp.link = p.link
            p.link = temp
            
    def insert_before(self,data,x):
        #If list is empty
        if self.start is None:
            print("List is empty")
            return
        
        #if x is in first node, then new node should be inserted before it
        if x == self.start.info:
            temp = node(data)
            temp.link = self.start
            self.start = temp
            return
        
        #Finding reference to predecessor of node containing x
        p = self.start
        while p.link is not None:
            if p.link.info == x:
                break
            p = p.link
            
        if p.link is None:
            print(x," not present in the list")
        else:
            temp = node(data)
            temp.link = p.link
            p.link = temp
            
    def insert_at_position(self,data,k):
        if k==1:
            temp = node(data)
            temp.link = self.start
            self.start = temp
            return
        
        p = self.start
        i = 1
        while i<k-1 and p is not None:
            p = p.link
            i += 1
            
        if p is None:
            print("You can only insert upto",i)
        else:
            temp = node(data)
            temp.link = p.link
            p.link = temp
            
###################DELETION#######################
    
    def delete_node(self,x):
        if self.start is None:
            print("List is empty")
            return
        
        #Deletion at first node
        if self.start.info == x:
            self.start = self.start.link
            return
        
        #Deletion in between or at the end
        p = self.start
        while p.link is not None:
            if p.link == x:
                break
            p = p.link
        if p.link is None:
            print("Element not in the list")
        else:
            p.link = p.link.link
            
    def delete_first_node(self):
        if self.start is None:
            return
        self.start = self.start.link
        
    def delete_last_node(self):
        if self.start is None:
            return
        if self.start.link is None:
            self.start = None
            return
        p = self.start
        while p.link.link is not None:
            p = p.link
        p.link = None
        
#Reversing the linked list   
    def reverse_list(self):
        prev = None
        p = self.start
        while p is not None:
            next = p.link
            p.link = prev
            prev = p
            p = next
        self.start = prev    
```

```python
list = SingleLinkedList()
print()
while True:
    print("1.Create list")
    print("2.Display list")
    print("3.Count the number of nodes")
    print("4.Search")
    print("5.Insert node in an empty/beginning of the list")
    print("6.Insert node at the end of the list")
    print("7.Insert node after a specified node")
    print("8.Insert node before a specified node")
    print("9.Insert node at a given position")
    print("10.Delete first node")
    print("11.Delete last node")
    print("12.Delete any node")
    print("13.Reverse the list")
    print("14.Quit")
    print()
    
    option = int(input("Enter a choice: "))
    
    if option == 1:
        list.create_list()
    elif option == 2:
        list.display_list()
    elif option == 3:
        list.count_nodes()
    elif option == 4:
        data = int(input("Enter the element to be searched: "))
        list.search(data)
    elif option == 5:
        data = int(input("Enter the element to be inserted: "))
        list.insert_at_beginning(data)
    elif option == 6:
        data = int(input("Enter the element to be inserted: "))
        list.insert_at_end(data)
    elif option == 7:
        data = int(input("Enter the element to be inserted: "))
        x = int(input("Enter the element after which to insert: "))
        list.insert_after(data,x)
    elif option == 8:
        data = int(input("Enter the element to be inserted: "))
        x = int(input("Enter the element before which to insert: "))
        list.insert_before(data,x)
    elif option == 9:
        data = int(input("Enter the element to be inserted: "))
        x = int(input("Enter the position at which to insert: "))
        list.insert_at_position(data,x)    
    elif option == 10:
        list.delete_first_node()
    elif option == 11:
        list.delete_last_node()
    elif option == 12:
        x = int(input("Enter the element to be deleted: "))
        list.delete_node(x)
    elif option == 13:
        list.reverse_list()
    elif option == 14:
        break
    else:
        print("Wrong option")
    print()
```
